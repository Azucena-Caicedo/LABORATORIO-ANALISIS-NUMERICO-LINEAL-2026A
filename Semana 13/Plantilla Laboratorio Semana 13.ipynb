# Método del Descenso Más Profundo (Steepest Descent)

## Idea principal

El método del descenso más profundo es un método iterativo para resolver un sistema lineal

$$
Ax=b,
$$

cuando la matriz \(A\) es **simétrica definida positiva (SDP)**.

En lugar de resolver directamente el sistema, se plantea un problema de optimización equivalente:

$$
\min \phi(x)=\frac12 x^T A x-b^T x.
$$

El mínimo de esta función corresponde exactamente a la solución del sistema lineal.

---

## Idea del método

En cada iteración se busca la dirección en la que la función disminuye más rápidamente.

Esa dirección es el **gradiente negativo** de la función objetivo:

$$
d_k=-\nabla \phi(x^{(k)}).
$$

Como

$$
\nabla\phi(x)=Ax-b,
$$

el gradiente negativo coincide con el residuo

$$
r^{(k)}=b-Ax^{(k)}.
$$

Por tanto, la dirección de búsqueda es

$$
d_k=r^{(k)}.
$$

---

## Actualización

La nueva aproximación se calcula mediante

$$
x^{(k+1)}=x^{(k)}+\alpha_k r^{(k)},
$$

donde $\alpha_k$ es el tamaño del paso elegido para minimizar la función sobre esa dirección.

---

## Ventajas

- Método sencillo de implementar.
- Requiere poca memoria.
- Garantiza convergencia cuando $A$ es simétrica definida positiva.

---

## Desventajas

- La convergencia suele ser lenta.
- Puede requerir muchas iteraciones.

---

# Método del Gradiente Conjugado (Conjugate Gradient)

## Idea principal

El método del gradiente conjugado también resuelve sistemas

$$
Ax=b,
$$

con $A$ simétrica definida positiva.

Su objetivo es acelerar el método del descenso más profundo.

---

## Idea del método

En lugar de utilizar únicamente el gradiente negativo como dirección de búsqueda, construye una sucesión de **direcciones conjugadas** (o **A-ortogonales**).

Estas direcciones evitan repetir información ya utilizada y permiten avanzar de forma más eficiente hacia la solución.

---

## Dirección de búsqueda

La primera dirección coincide con el residuo

$$
p^{(0)}=r^{(0)}.
$$

Las siguientes direcciones se calculan combinando el nuevo residuo con la dirección anterior, manteniendo la propiedad de conjugación.

---

## Actualización

La aproximación se actualiza mediante

$$
x^{(k+1)}=x^{(k)}+\alpha_k p^{(k)},
$$

donde $p^{(k)}$ es la dirección conjugada.

---

## Ventajas

- Converge mucho más rápido que el descenso más profundo.
- Aprovecha la información de iteraciones anteriores.
- Es uno de los métodos más eficientes para matrices simétricas definidas positivas de gran tamaño.

---

## Comparación

| Descenso más profundo | Gradiente conjugado |
|------------------------|---------------------|
| Usa el gradiente negativo. | Usa direcciones conjugadas. |
| Método simple. | Método más eficiente. |
| Convergencia lenta. | Convergencia mucho más rápida. |
| Puede necesitar muchas iteraciones. | Generalmente necesita muchas menos iteraciones. |
| Base para entender métodos más avanzados. | Método ampliamente utilizado para sistemas lineales grandes. |

---

## Idea clave

- **Descenso más profundo:** siempre avanza en la dirección de máximo descenso (gradiente negativo).
- **Gradiente conjugado:** construye direcciones más inteligentes para evitar repetir trabajo y acelerar la convergencia.

# Ejemplo: Método del Descenso Más Profundo

Resolver el sistema

$$
Ax=b,
$$

donde

$$
A=
\begin{bmatrix}
4&1\\
1&3
\end{bmatrix},
\qquad
b=
\begin{bmatrix}
1\\
2
\end{bmatrix}.
$$

La matriz $A$ es simétrica definida positiva, por lo que podemos aplicar el método.

---

## Paso 1. Plantear el problema de optimización

Resolver el sistema es equivalente a minimizar

$$
\phi(x)=\frac12x^TAx-b^Tx.
$$

---

## Paso 2. Elegir una aproximación inicial

Tomamos

$$
x^{(0)}=
\begin{bmatrix}
0\\
0
\end{bmatrix}.
$$

---

## Paso 3. Calcular el residuo

El residuo es

$$
r^{(0)}=b-Ax^{(0)}.
$$

Como

$$
Ax^{(0)}=
\begin{bmatrix}
0\\
0
\end{bmatrix},
$$

se obtiene

$$
r^{(0)}=
\begin{bmatrix}
1\\
2
\end{bmatrix}.
$$

Esta será la dirección de búsqueda.

---

## Paso 4. Calcular el tamaño del paso

En los apuntes,

$$
\alpha_k=
\frac{r_k^Tr_k}
{r_k^TAr_k}.
$$

Calculamos

$$
Ar^{(0)}=
\begin{bmatrix}
6\\
7
\end{bmatrix}.
$$

Luego,

$$
r^{(0)T}r^{(0)}=1^2+2^2=5,
$$

y

$$
r^{(0)T}Ar^{(0)}
=
1(6)+2(7)=20.
$$

Por tanto,

$$
\boxed{\alpha_0=\frac{5}{20}=0.25.}
$$

---

## Paso 5. Actualizar la solución

Usamos

$$
x^{(1)}
=
x^{(0)}
+
\alpha_0r^{(0)}.
$$

Entonces

$$
x^{(1)}
=
\begin{bmatrix}
0\\
0
\end{bmatrix}
+
0.25
\begin{bmatrix}
1\\
2
\end{bmatrix}
=
\boxed{
\begin{bmatrix}
0.25\\
0.50
\end{bmatrix}}.
$$

---

## Paso 6. Calcular el nuevo residuo

Primero,

$$
Ax^{(1)}
=
\begin{bmatrix}
1.5\\
1.75
\end{bmatrix}.
$$

Luego,

$$
r^{(1)}
=
b-Ax^{(1)}
=
\boxed{
\begin{bmatrix}
-0.50\\
0.25
\end{bmatrix}}.
$$

Se continúa repitiendo los mismos pasos hasta que el residuo sea suficientemente pequeño.

---

# Ejemplo: Método del Gradiente Conjugado

Resolvemos exactamente el mismo sistema

$$
A=
\begin{bmatrix}
4&1\\
1&3
\end{bmatrix},
\qquad
b=
\begin{bmatrix}
1\\
2
\end{bmatrix}.
$$

---

## Paso 1. Aproximación inicial

Tomamos

$$
x^{(0)}
=
\begin{bmatrix}
0\\
0
\end{bmatrix}.
$$

---

## Paso 2. Residuo inicial

$$
r^{(0)}
=
b-Ax^{(0)}
=
\begin{bmatrix}
1\\
2
\end{bmatrix}.
$$

---

## Paso 3. Primera dirección conjugada

La primera dirección siempre es

$$
p^{(0)}
=
r^{(0)}
=
\begin{bmatrix}
1\\
2
\end{bmatrix}.
$$

---

## Paso 4. Calcular el tamaño del paso

Se utiliza la misma expresión

$$
\alpha_0=
\frac{r_0^Tr_0}
{p_0^TAp_0}.
$$

Como

$$
p^{(0)}=r^{(0)},
$$

obtenemos

$$
\alpha_0=\frac{5}{20}=0.25.
$$

---

## Paso 5. Actualizar la solución

$$
x^{(1)}
=
x^{(0)}
+
\alpha_0p^{(0)}.
$$

Así,

$$
x^{(1)}
=
\boxed{
\begin{bmatrix}
0.25\\
0.50
\end{bmatrix}}.
$$

---

## Paso 6. Calcular el nuevo residuo

$$
r^{(1)}
=
b-Ax^{(1)}
=
\boxed{
\begin{bmatrix}
-0.50\\
0.25
\end{bmatrix}}.
$$

---

## Paso 7. Calcular el parámetro

Se calcula

$$
\beta_0=
\frac{r_1^Tr_1}
{r_0^Tr_0}.
$$

Como

$$
r_1^Tr_1
=
(-0.5)^2+(0.25)^2
=
0.3125,
$$

y

$$
r_0^Tr_0=5,
$$

obtenemos

$$
\boxed{\beta_0=0.0625.}
$$

---

## Paso 8. Nueva dirección conjugada

Ahora ya no se usa solamente el residuo.

Se calcula

$$
p^{(1)}
=
r^{(1)}
+
\beta_0p^{(0)}.
$$

Por tanto,

$$
p^{(1)}
=
\begin{bmatrix}
-0.50\\
0.25
\end{bmatrix}
+
0.0625
\begin{bmatrix}
1\\
2
\end{bmatrix}
=
\boxed{
\begin{bmatrix}
-0.4375\\
0.375
\end{bmatrix}}.
$$

A partir de aquí el algoritmo continúa calculando un nuevo valor de $\alpha$, actualizando $x$, calculando el nuevo residuo y construyendo una nueva dirección conjugada.

---

| Puede necesitar muchas iteraciones. | Generalmente converge en muchas menos iteraciones. |
| Es más sencillo. | Es más eficiente para matrices simétricas definidas positivas. |

```python
import numpy as np

# ==========================================================
# FUNCIÓN: MÉTODO DEL GRADIENTE CONJUGADO
# ==========================================================

def gradiente_conjugado(A, b, x0, tol=1e-8, max_iter=20):

    x = x0.copy()                               # Copia del valor inicial

    r = b - A @ x                               # Residuo inicial

    p = r.copy()                                # Primera dirección conjugada

    for k in range(max_iter):

        error = np.linalg.norm(r)               # Norma del residuo

        if error < tol:                         # Criterio de parada
            break

        alpha = (r @ r) / (p @ A @ p)           # Tamaño del paso

        x = x + alpha * p                       # Actualiza la solución

        #COMPLETA EN CLASE                      # Nuevo residuo

        #COMPLETA EN CLASE                      # Calcula beta

        #COMPLETA EN CLASE                       # Nueva dirección conjugada

        #COMPLETA EN CLASE                             # Actualiza el residuo

        print(f"Iteración {k+1}")
        print("x =", x)
        print("Residuo =", error)
        print()

    return x, k+1, error                        # Devuelve solución, iteraciones y error


# ==========================================================
# DATOS DEL PROBLEMA
# ==========================================================

A = np.array([[4., 1.],
              [1., 3.]])

b = np.array([1., 2.])

x0 = np.array([0., 0.])


# ==========================================================
# EJECUTAR EL MÉTODO
# ==========================================================

x_cg, it_cg, err_cg = gradiente_conjugado(A, b, x0)

print("Solución aproximada:")
print(x_cg)
```

## Conclusión

El **Gradiente Conjugado** puede verse como una mejora del **Descenso Más Profundo**.

Ambos comienzan exactamente igual:

1. Calcular el residuo.
2. Calcular el tamaño del paso.
3. Actualizar la solución.

La diferencia es que el **Descenso Más Profundo** siempre utiliza el residuo como dirección de búsqueda, mientras que el **Gradiente Conjugado** construye una nueva dirección combinando el residuo actual con la dirección anterior mediante el parámetro

$$
\beta_k,
$$

lo que evita repetir información y acelera la convergencia del método.
