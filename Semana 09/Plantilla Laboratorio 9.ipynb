El método de Jacobi es convergente para cualquier valor inicial si $A$ es una matriz diagonal dominante (por filas o columnas).  
El valor absoluto del elemento diagonal debe ser mayor o igual que la suma de los demás elementos de esa fila.

# Métodos Iterativos: Jacobi y Gauss-Seidel

Queremos resolver el sistema lineal:

$$
A\mathbf{x}=\mathbf{b}
$$

La idea de los métodos iterativos es escribir el sistema en la forma:

$$
\mathbf{x}=B\mathbf{x}+\mathbf{d}
$$

donde:

- $B$ = matriz de iteración
- $\mathbf{d}$ = vector constante

Partiendo de una aproximación inicial:

$$
\mathbf{x}^{(1)}
$$

se generan nuevas aproximaciones usando:

$$
\mathbf{x}^{(k+1)}=
B\mathbf{x}^{(k)}+\mathbf{d}
$$

para:

$$
k=1,2,3,\dots
$$

Esto produce una sucesión:

$$
\mathbf{x}^{(1)},
\mathbf{x}^{(2)},
\mathbf{x}^{(3)},
\dots
$$

que idealmente converge a la solución exacta:

$$
\mathbf{x}
$$

---

## Idea general

Los métodos iterativos responden dos preguntas:

### (a) ¿Cómo escribir?

$$
A\mathbf{x}=b
$$

en la forma:

$$
\mathbf{x}=B\mathbf{x}+d
$$

### (b) ¿Cuándo converge?

Es decir:

- ¿Qué aproximación inicial usar?
- ¿Bajo qué condiciones converge?

---

## Métodos estudiados

Existen tres métodos iterativos clásicos:

- Jacobi
- Gauss-Seidel
- SOR (Successive Over Relaxation)

La diferencia principal entre ellos es:

> **Cómo construyen la matriz de iteración $B$ y el vector $d$.**

# Fórmula general del método de Jacobi

Partiendo del sistema:

$$
A\mathbf{x}=\mathbf{b}
$$

cada ecuación:

$$
a_{i1}x_1+a_{i2}x_2+\cdots+a_{ii}x_i+\cdots+a_{in}x_n=b_i
$$

se despeja respecto a:

$$
x_i
$$

obteniéndose:

$$
x_i^{(k+1)}
=
\frac{
b_i-
\sum_{j\neq i}^{n}
a_{ij}x_j^{(k)}
}{
a_{ii}
}
\qquad i=1,2,\dots,n
$$

Observación:

> Jacobi utiliza únicamente valores de la iteración anterior.

---

# Fórmula general del método de Gauss-Seidel

El método de Gauss-Seidel modifica Jacobi reutilizando inmediatamente los nuevos valores calculados.

La fórmula general es:

$$
x_i^{(k+1)}
=
\frac{
1
}{
a_{ii}
}
\left(
b_i
-
\sum_{j=1}^{i-1}
a_{ij}x_j^{(k+1)}
-
\sum_{j=i+1}^{n}
a_{ij}x_j^{(k)}
\right)
$$

para:

$$
i=1,2,\dots,n
$$

Observación:

> Gauss-Seidel utiliza inmediatamente los valores recién calculados dentro de la misma iteración.

---

Jacobi:

$$

$$

\rightarrow
x_2^{(k+1)}
\rightarrow
x_3^{(k+1)}
\rightarrow
\cdots
$$

> Gauss-Seidel normalmente converge en menos iteraciones.

# Conclusiones: Rapidez y Convergencia

## Rapidez de convergencia

Observamos que:

- **Jacobi necesita más iteraciones** para aproximarse a la solución.
- **Gauss-Seidel converge más rápidamente.**

### ¿Por qué ocurre esto?

### Jacobi

En Jacobi, toda la nueva iteración se calcula usando únicamente los valores anteriores:

$$
x^{(k+1)}=Bx^{(k)}+d
$$

Es decir:

> **espera terminar toda la iteración antes de utilizar nuevos valores.**

Por ejemplo:

Si estamos calculando:

$$
x_1,x_2,x_3
$$

todos utilizan información de:

$$
x^{(k)}
$$

---

### Gauss-Seidel

Gauss-Seidel funciona diferente.

Apenas calcula un nuevo valor:

$$
x_1^{(k+1)}
$$

lo utiliza inmediatamente para calcular:

$$
x_2^{(k+1)}
$$

y luego:

$$
x_3^{(k+1)}
$$

Es decir:

> **usa información más reciente dentro de la misma iteración.**

Visualmente:

Jacobi:

$$
(x^{(k)})\rightarrow(x^{(k+1)})
$$

Gauss-Seidel:

$$
x_1^{(k+1)}
\rightarrow
x_2^{(k+1)}
\rightarrow
x_3^{(k+1)}
$$

Por esta razón normalmente necesita menos iteraciones.
