## Epsilon de máquina

### Definición

El **epsilon de máquina** es el número más pequeño $\varepsilon$ tal que:

$$
1 + \varepsilon \neq 1
$$

en aritmética de punto flotante.

---

### Valor típico

En doble precisión ($\text{float64}$), se tiene:

$$
\varepsilon \approx 2.22 \times 10^{-16}
$$

---



float(64) Un número almacenado en 64 bits (8 bytes)

````python
np.float64
````

Podemos imprimir el epsilon de maquina

````python
eps=np.finfo(np.float64).eps
print(eps)
````

#### 3. Evitar inestabilidad numérica

Para evitar divisiones peligrosas-

**Ejemplo**: Considere la expresión:

$$
\frac{1}{x}
$$

Si $x$ es muy pequeño (cercano a cero), el resultado puede ser numéricamente inestable o incluso producir errores.

---

### ⚠️ Problema

```python
x = 1e-20
print(1/x)
```

Esto produce un número extremadamente grande, lo cual puede generar:

* inestabilidad numérica
* overflow
* pérdida de precisión

**Overflow** ocurre cuando un cálculo produce un número tan grande que no puede representarse en la computadora.

En ese caso, el resultado se vuelve:

$$
\infty \ (\text{infinito})
$$

o genera un error numérico.

````python
#Solución

eps=np.finfo(np.float64).eps
x=1e-20
den=max(x,eps)
print(den)
print(1/den)
````

### Ejemplo: precisión de máquina

Considere la expresión:

$$
1 - (1 + \varepsilon)
$$

Desde el punto de vista matemático:

$$
1 - (1 + \varepsilon) = -\varepsilon
$$

---


````python
print(1-(1+eps))
````

### Interpretación

El resultado es aproximadamente:

$$
-\varepsilon
$$

Esto muestra que la computadora **sí distingue entre $1$ y $1 + \varepsilon$**, es decir:

$$
1 + \varepsilon \neq 1
$$

### ⚠️ Ejemplo: errores de redondeo

Considere la suma:

$$
0.1 + 0.2
$$

Desde el punto de vista matemático:

$$
0.1 + 0.2 = 0.3
$$

---

### ⚙️ Implementación en Python

```python
print(0.1 + 0.2)

````python
print(0.1 + 0.2)
````

### Interpretación

El resultado no es exactamente $0.3$ debido a **errores de redondeo** en la representación de números en punto flotante.

Esto ocurre porque números como $0.1$ y $0.2$ no pueden representarse exactamente en base binaria.

---

### Idea clave

Los errores de redondeo:

- son inevitables en computación  
- dependen de la precisión de la máquina  
- afectan operaciones con números decimales
