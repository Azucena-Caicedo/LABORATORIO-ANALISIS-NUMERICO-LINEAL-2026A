## Objetivo general

Comprender cómo el precondicionamiento y el análisis espectral permiten acelerar la convergencia de los métodos iterativos para resolver sistemas lineales, así como introducir las bases teóricas que conducen a métodos más eficientes como el Gradiente Conjugado.

# Matrices de interacción y precondicionamiento

## Idea principal

Muchos métodos iterativos para resolver

$$
Ax=b
$$

pueden interpretarse como la resolución de un sistema equivalente

$$
M^{-1}Ax=M^{-1}b,
$$

donde $M$ es una matriz llamada **precondicionador**.

La idea es elegir $M$ de manera que el nuevo sistema sea más fácil de resolver y los métodos iterativos converjan más rápido.

---

## ¿Por qué usar un precondicionador?

Un buen precondicionador puede:

- Reducir el número de iteraciones.
- Mejorar el número de condición de la matriz.
- Acelerar la convergencia.
- Hacer más eficientes métodos como Gradiente Conjugado.

---

## Interpretación

La idea principal es que muchos métodos iterativos (Jacobi, Gauss-Seidel, SOR, etc.) no trabajan directamente con la matriz $A$, sino con una matriz más sencilla $M$.

Partimos del sistema

$$
Ax=b.
$$

Supongamos que descomponemos la matriz $A$ como

$$
A=M-N.
$$

Por ejemplo, en Jacobi:

$$
A=D-(L+U),
$$

donde:

- $D$: diagonal de $A$,
- $L$: parte triangular inferior,
- $U$: parte triangular superior.

---

### Paso 1: sustituir la descomposición

Reemplazamos $A=M-N$ en el sistema original:

$$
(M-N)x=b.
$$

Reordenando,

$$
Mx=Nx+b.
$$

---

### Paso 2: despejar $x$

Multiplicamos por $M^{-1}$:

$$
x=M^{-1}Nx+M^{-1}b.
$$

Esta ecuación tiene exactamente la forma de un método iterativo:

$$
x^{(k+1)}
=
M^{-1}N\,x^{(k)}
+
M^{-1}b.
$$

Por comparación con

$$
x^{(k+1)}
=
Bx^{(k)}+c,
$$

obtenemos

$$
B=M^{-1}N,
\qquad
c=M^{-1}b.
$$

---

### Paso 3: obtener otra expresión para $B$

Como

$$
A=M-N,
$$

entonces

$$
N=M-A.
$$

Sustituyendo en

$$
B=M^{-1}N,
$$

tenemos

$$
B=M^{-1}(M-A).
$$

Distribuyendo,

$$
B=M^{-1}M-M^{-1}A.
$$

Como

$$
M^{-1}M=I,
$$

resulta

$$
B=I-M^{-1}A.
$$

---

### ¿Qué significa esto?

La matriz que realmente controla la convergencia del método iterativo es

$$
B=I-M^{-1}A.
$$

Por tanto, el comportamiento del método depende de la matriz

$$
M^{-1}A.
$$

Si elegimos un buen precondicionador $M$ el método convergerá más rápido.

---

### Ejemplo sencillo

Sea

$$
A=
\begin{pmatrix}
4 & 1\\
1 & 3
\end{pmatrix}.
$$

En Jacobi tomamos

$$
M=D=
\begin{pmatrix}
4 & 0\\
0 & 3
\end{pmatrix}.
$$

Entonces

$$
M^{-1}
=
\begin{pmatrix}
1/4 & 0\\
0 & 1/3
\end{pmatrix}.
$$

Calculamos

$$
M^{-1}A
=
\begin{pmatrix}
1 & 0.25\\
0.333 & 1
\end{pmatrix}.
$$

En lugar de trabajar directamente con $A$, el método trabaja con $M^{-1}A$.

Por eso se dice que estamos resolviendo el sistema precondicionado

$$
M^{-1}Ax=M^{-1}b.
$$

La idea es que $M^{-1}A$ sea "más fácil" para el método iterativo que la matriz original $A$.
# Precondicionadores clásicos

## Precondicionador de Jacobi

Utiliza únicamente la diagonal de la matriz:

$$
M_J=D.
$$

---

### Ejemplo

Sea

$$
A=
\begin{pmatrix}
4 & 1\\
1 & 3
\end{pmatrix}.
$$

La diagonal es

$$
D=
\begin{pmatrix}
4 & 0\\
0 & 3
\end{pmatrix}.
$$

Por tanto,

$$
M_J=D.
$$

---

## Precondicionador de Gauss-Seidel

Utiliza la diagonal y la parte triangular inferior:

$$
M_{GS}=D+L.
$$

---

### Ejemplo

Sea

$$
A=
\begin{pmatrix}
4 & 1\\
2 & 3
\end{pmatrix}.
$$

Entonces

$$
D=
\begin{pmatrix}
4 & 0\\
0 & 3
\end{pmatrix},
\qquad
L=
\begin{pmatrix}
0 & 0\\
2 & 0
\end{pmatrix}.
$$

Por tanto,

$$
M_{GS}
=
D+L
=
\begin{pmatrix}
4 & 0\\
2 & 3
\end{pmatrix}.
$$

---

## Precondicionador SOR

Incorpora un parámetro de relajación $\omega$:

$$
M_{SOR}
=
\frac{1}{\omega}(D+\omega L).
$$

---

### Ejemplo

Si

$$
\omega=1.2,
$$

entonces

$$
M_{SOR}
=
\frac{1}{1.2}
(D+1.2L).
$$

La elección adecuada de $\omega$ puede acelerar la convergencia.

# Teorema de Gershgorin

## Idea principal

Permite estimar dónde se encuentran los valores propios de una matriz sin calcularlos explícitamente.

Todo valor propio $\lambda$ de una matriz $A$ pertenece a al menos uno de los discos

$$
|\lambda-a_{ii}|
\le
\sum_{j\neq i}|a_{ij}|.
$$

---
## Ejemplo

Sea

$$
A=
\begin{pmatrix}
4 & 1\\
2 & 5
\end{pmatrix}.
$$

Para aplicar el teorema de Gershgorin, se construye un disco por cada fila de la matriz.

### Primer disco

La primera fila es

$$
(4,\;1).
$$

El centro del disco es el elemento diagonal:

$$
a_{11}=4.
$$

El radio es la suma de los valores absolutos de los elementos fuera de la diagonal en esa fila:

$$
R_1
=
|1|
=
1.
$$

Por tanto, el primer disco de Gershgorin es

$$
|\lambda-4|\le 1.
$$

Esto significa que todos los valores propios asociados a este disco deben encontrarse dentro del intervalo

$$
[4-1,\;4+1]
=
[3,\;5].
$$

---

### Segundo disco

La segunda fila es

$$
(2,\;5).
$$

El centro del disco es el elemento diagonal:

$$
a_{22}=5.
$$

El radio es la suma de los valores absolutos de los elementos fuera de la diagonal en esa fila:

$$
R_2
=
|2|
=
2.
$$

Por tanto, el segundo disco de Gershgorin es

$$
|\lambda-5|\le 2.
$$

Esto significa que los valores propios contenidos en este disco deben encontrarse dentro del intervalo

$$
[5-2,\;5+2]
=
[3,\;7].
$$

---

### Conclusión

Los valores propios de la matriz

$$
A=
\begin{pmatrix}
4 & 1\\
2 & 5
\end{pmatrix}
$$

deben encontrarse dentro de la unión de los discos de Gershgorin:

$$
|\lambda-4|\le 1,
\qquad
|\lambda-5|\le 2.
$$

o equivalentemente dentro del intervalo

$$
[3,7].
$$

Por lo tanto,

$$
\lambda(A)\subset [3,7].
$$

## Interpretación

Los valores propios deben encontrarse dentro de alguno de estos discos.

Esto permite:

- Estimar la distribución espectral.
- Analizar convergencia de métodos iterativos.
- Evaluar la calidad de un precondicionador.
